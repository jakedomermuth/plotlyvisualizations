```python
import plotly.express as px
import pandas as pd
t_data = pd.read_csv('/content/drive/MyDrive/Tornados 2023-2.csv')
t_data2 = t_data.query("st in ['TN', 'FL', 'VA', 'TX', 'CA']")
fig = px.pie(t_data2, values='loss', names='st', color_discrete_sequence=px.colors.sequential.RdBu,  title='Loss By State')
fig.show()
```

```python
import plotly.express as px
import pandas as pd
t_data = pd.read_csv('/content/drive/MyDrive/Tornados 2023-2.csv')
t_data2 = t_data.query("st in ['TN', 'FL', 'VA', 'TX', 'CA']")
t_data3 = t_data2.groupby(['st'])['inj'].mean().reset_index()
fig = px.pie(t_data3, values='inj', names='st', color_discrete_sequence=px.colors.sequential.RdBu,  title='Average inj By State')
fig.show()
```

```python
import plotly.express as px
import pandas as pd
t_data = pd.read_csv('/content/drive/MyDrive/Tornados 2023-2.csv')
total_tornados = t_data.groupby(['st'])['om'].count().reset_index()
fig = px.choropleth(total_tornados,
                    locations='st',
                    color='om',
                    color_continuous_scale='spectral_r',
                    hover_name='st',
                    locationmode='USA-states',
                    labels={'om':' Count Of Tornados', 'st': 'State'},
                    scope='usa')
fig.add_scattergeo(
    locations=t_data['st'],
    locationmode='USA-states',
    text=t_data['st'],
    hoverinfo='none',
    mode='text')
fig.show()
```
